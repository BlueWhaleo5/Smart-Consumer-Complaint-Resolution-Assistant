```python
import os
import sys
from dotenv import load_dotenv

# Thêm đường dẫn src để import các module local
sys.path.append(os.path.abspath("../"))

from src.rag_pipeline.retriever import ComplaintRetriever
from src.rag_pipeline.qa_chain import QAChainManager

load_dotenv("../.env")
print("Môi trường đã sẵn sàng.")
```

Thử nghiệm Retrieval (Truy xuất)  
Kiểm tra xem hệ thống có tìm đúng kiến thức từ `SCCRA.pdf` hoặc SQLite không

```python
retriever = ComplaintRetriever()
query = "How does the project handle high priority complaints?"
context, sources = retriever.get_context(query, k=3)

print(f"--- Context tìm được ---\n{context[:500]}...")
print(f"\n--- Nguồn ---\n{sources}")
```

Thử nghiệm QA Chain (Toàn bộ luồng)  
Gửi câu hỏi cho LLM kèm theo context đã tìm thấy

```python
qa_manager = QAChainManager(use_local=False) # Sử dụng Gemini
question = "Summary the financial products covered in this project."
answer, sources = qa_manager.get_answer(question)

print(f"AI ANSWER:\n{answer}")
print(f"\nSOURCES: {sources}")
```
